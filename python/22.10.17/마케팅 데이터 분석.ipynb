### 기초정보

```python
imp = 10000 #노출수
clk = 100 #클릭수
conv = 10 #구매수
cost = 100000 #광고비용
```

### ctr과 cpm

```python
# 클릭율
ctr = clk/imp*100
ctr
```

```python
# cpm
cpm = cost/imp*1000
cpm
```

###cpc와 cpa

```python
# cpc
cpc = cost/clk
cpc
```

```python
# cpa
cpa = cost/conv
cpa
```
