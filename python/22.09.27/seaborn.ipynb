### seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import numpy as np

x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 4*np.sin(x+1.5)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
sns.set_style("white")
plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
sns.despine(); plt.show()
```

```python
sns.set_style("dark")
plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
sns.set_style("white")
plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
sns.despine(offset=10)
plt.show()
```

```python
tips = sns.load_dataset("tips")
tips.head(5)
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=tips["total_bill"])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x="day", y="total_bill", data=tips, color=".5")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips, height=7)
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, height=7)
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights.head(5)
```

```python
flights = flights.pivot("month", "year", "passengers")
flights.head(5)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt="d")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt="d", cmap="YlGnBu")
plt.show()
```

### Box plot

```python
samples = [1, 7, 9, 16, 36, 39, 45, 45, 46, 48, 51, 100, 101]
tmp_y = [1]*len(samples)
```

```python
tmp_y
```

```python
plt.figure(figsize=(12,4))
plt.scatter(samples, tmp_y)
plt.grid()
plt.show()
```

```python
np.median(samples)
```

```python
np.percentile(samples,25)
```

```python
np.percentile(samples,75)
```

```python
np.percentile(samples,75) - np.percentile(samples,25)
```

```python
iqr = np.percentile(samples,75) - np.percentile(samples,25)
iqr*1.5
```

```python
q1 = np.percentile(samples,25)
q2 = np.median(samples)
q3 = np.percentile(samples,75)
upper_fence = q3 + iqr*1.5
lower_fence = q1 - iqr*1.5
```

```python
plt.figure(figsize=(12,4))
plt.scatter(samples, tmp_y)
plt.axvline(x=q1, color='black')
plt.axvline(x=q2, color='red')
plt.axvline(x=q3, color='black')
plt.axvline(x=upper_fence, color='black', ls='dashed')
plt.axvline(x=lower_fence, color='black', ls='dashed')
plt.grid()
plt.show()
```
