#### **3. URL**

- Client -> url -> Server
- https://en.wikipedia.org/wiki/URL
- URL: Uniform Resource Locator (URL)
- ```URL scheme ":" ["//" authority] path ["?" query] ["#" fragment]```

````python
img = Image.open('img/URI_syntax_diagram.svg.png')
img
````

> #### **네이버 스포츠 뉴스 URL 예시**
```
https://sports.news.naver.com/news?oid=109&aid=0004707260
```

- https -> scheme 
- sports.news.naver.com -> authority(subdomain, domain)
- news -> path 
- ?oid=109&aid=0004707260 -> query 
```
