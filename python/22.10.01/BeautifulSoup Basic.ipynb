## 2-2. bs4 예제 네이버 금융
- find, find_all

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://finance.naver.com/marketindex/'
response = requests.get(url)
response = BeautifulSoup(response.content, 'html.parser')
```

```python
soup.select('#exchangeList > li')
```

```python
len(soup.select('#exchangeList > li'))
```

```python
# 전체 데이터 수집
exchangeList = soup.select('#exchangeList > li')
exchangeList[0].text
```

```python
# 가져온 전체 리스트에서 세부 데이터 추출
# 나라 이름 , 환율 , 상승/하락, 상승/하락 수치

title = exchangeList[0].select_one('.blind').text
updown = exchangeList[0].select_one('.head_info > .blind').text
change = exchangeList[0].select_one('.change').text
exchange = exchangeList[0].select_one('.value').text

# title, exchange, updown, change
```

```python
exchangeList[0].select_one('#exchangeList > li.on > a.head.usd > h3 > span') 
```

```python
exchangeList[0].select_one('.blind') 
```

```python
exchangeList[0].select_one('.blind').text
```

```python
exchangeList[0].select_one('.blind').text, exchangeList[0].select_one('.value').text
```

```python
# 클래스 이름이 같을 때
exchangeList[0].select_one('.head_info > .blind').text
```

```python
exchangeList[0].select_one('.change').text
```

```python
# 전체 데이터 수집, 데이터프레임으로 만들기
import pandas as pd

datas = []

for item in exchangeList:
    data = {
        "title": item.select_one('.blind').text
        "exchange": item.select_one('.value').text
        "change": item.select_one('.change').text
        "updown": item.select_one('.head_info > .blind').text
    }
    datas.append(data)
    
df = pd.DataFrame(datas)
df
```
