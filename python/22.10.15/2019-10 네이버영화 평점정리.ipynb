## 날짜와 제목, 평점을 저장할 빈 리스트

```python
import time
movie_date = []
movie_name = []
movie_point = []
```

## 한 페이지에서 만들어지는 정보르 extend

```python
for today in date:
    html = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date={date}"
    response = urlopen(html.format(date = today.strftime('%Y%m%d')))
    soup = BeautifulSoup(response, "html.parser")
    
    movie_date.extend([today]*len(soup.find_all('td', 'point')))
    movie_name.extend([each.a.string for each in soup.find_all('div', 'tit5')])
    movie_point.extend([each.string for each in soup.find_all('td', 'point')])
    
    print(str(today))
    time.sleep(1)
```

## 세 리스트의 크기 문제없음

```python
len(movie_date), len(movie_name), len(movie_point),
```

## pandas DataFrame 으로 생성

```python
movie = pd.DataFrame({'date':movie_date, 'name':movie_name, 'point':movie_point})
movie.head()
```

## 이상 없는지 info로 조회

```python
movie.info()
```

## 정렬 기능을 사용하기 위해 point를 float형으로

```python
movie['point'] = movie['point'].astype(float)
movie.info()
```

## 수시로 저장

```python
movie.to_csv("./naver_movie.csv", sep=",", encoding="utf-8")
```

```python
import numpy as np
import pandas as pd

movie = pd.read_csv("./naver_movie.csv", index_col=0)
movie.head()
```

## 이름으로 pivot_table

```python
movie_unique = pd.pivot_table(movie, index=['name'], aggfunc=np.sum)
movie_unique.head()
```

## 10월 한달간 평점의 합계

```python
movie_best = movie_unique.sort_values(by='point', ascending=False)
movie_best.head(10)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,8))
plt.plot(movie.query('name == ["조커"]')['date'],
         movie.query('name == ["조커"]')['point'])
plt.xticks('')
plt.ylim((5,10))
plt.grid()
plt.show()
```
